```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Carregar a imagem colorida
img = mpimg.imread('penguim.jpg')  # Imagem no formato [altura, largura, 3]

# 1. Converter para escala de cinza
def rgb_to_gray(rgb_img):

#  Extrai os canais de cor:
#  r: Canal Vermelho (ndice 0)
#  g: Canal Verde (ndice 1)
#  b: Canal Azul (ndice 2)

    r, g, b = rgb_img[:, :, 0], rgb_img[:, :, 1], rgb_img[:, :, 2]

#  Aplica a frmula de luminosidade padro NTSC para converso
    gray_img = 0.299 * r + 0.587 * g + 0.114 * b
    return gray_img.astype(np.uint8)

gray_img = rgb_to_gray(img)

# 2. Binarizar a imagem
def binarize(gray_img, threshold=160):
    # Aplica um limiar para binarizao
    # Executa a binarizao com limiar=160 e armazena em binary_img
    binary_img = np.zeros_like(gray_img)
    # Define o limiar
    # Pixels acima do limiar se tornam brancos (255), abaixo se tornam pretos
    binary_img[gray_img >= threshold] = 255
    return binary_img

binary_img = binarize(gray_img, threshold=160)

# Exibir os resultados
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')

# cmap='gray' -> Exibe a imagem em tons de cinza

plt.subplot(1, 3, 2)
plt.imshow(gray_img, cmap='gray')
plt.title('Escala de Cinza')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(binary_img, cmap='gray')
plt.title('Binria (160)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
